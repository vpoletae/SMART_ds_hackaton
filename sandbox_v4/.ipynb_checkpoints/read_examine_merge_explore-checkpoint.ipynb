### Explore & Clean up, Preliminary feature engineering

```python
labeled_merged = df_joint.copy()
```

```python
labeled_merged.columns
```

#### Remove special info (identifiers)

```python
labeled_merged.drop(columns=['label_id', 'label_form_id', 'label_user_id', 'label_goal_id', 'label_num', 'goal_form_id',
                      'goal_form_num', 'goal_id'], inplace=True)
```

```python
labeled_merged.info()
```

```python
labeled_merged.isnull().sum()
```

#### Label cols fillna

```python
labeled_merged['label_attractor_health'].fillna(0, inplace=True)
```

```python
labeled_merged['label_questions'].fillna(0, inplace=True)
```

```python
labeled_merged['label_custom_question'].fillna(0, inplace=True)
```

#### Personality info

Can't be filled correctly. Dangerous to drop. 'Undefined' is also an info - some sort of a sign

#####  - Sex

```python
labeled_merged['goal_sex'].fillna('Not mentioned', inplace=True)
```

```python
labeled_merged['goal_sex'].value_counts()
```

```python
valid_gender = ['Женщина', 'Мужчина', 'Not mentioned']
```

Can drop 3 lines

```python
labeled_merged = labeled_merged[labeled_merged['goal_sex'].isin(valid_gender) == True]
```

##### - Age

```python
labeled_merged['goal_age'].fillna(0, inplace=True)
```

```python
labeled_merged['goal_age'].astype('int').hist()
```

##### - City

```python
labeled_merged['goal_city'].fillna('Not mentioned', inplace=True)
```

```python
labeled_merged['goal_city'].value_counts()
```

##### - Activity

```python
labeled_merged['goal_activity'].fillna('Not mentioned', inplace=True)
```

##### - Job

```python
labeled_merged['goal_job'].fillna('Not mentioned', inplace=True)
```

#### Time & Obstackles

```python
labeled_merged['goal_time'].value_counts()
```

```python
values = labeled_merged['goal_time'].value_counts().keys().tolist()
counts = labeled_merged['goal_time'].value_counts().tolist()
```

```python
plt.hist(counts)
```

```python
unique_time_vals = []
for i in range(0, len(values)):
    if counts[i] == 1:
        unique_time_vals.append(values[i])
```

```python
unique_time_vals
```

```python
len(unique_time_vals)
```

First steps spoiled domain, obstackles and time

```python
labeled_merged['space'] = ' '
```

#### Goal type

```python
labeled_merged['goal_type'].head()
```

```python
len(labeled_merged['goal_type'].unique())
```

```python
labeled_merged['goal_type'].value_counts()
```

```python
from collections import defaultdict
```

```python
goal_type_split_unique = defaultdict(int)
for goal_type in list(labeled_merged['goal_type']):
    goal_type_split = str(goal_type).split(',')
    for type_part in goal_type_split:
        goal_type_split_unique[type_part.lstrip()] += 1
```

```python
goal_type_split_unique
```

Obviously, there are only several types of goal, the rest ones are or trash, or shifted cols

```python
labeled_merged['goal_type'].fillna('Not mentioned', inplace=True)
```
