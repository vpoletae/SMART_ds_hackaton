Cleanup pipeline:
1. lower
2. remove special symbols
3. lemmatize
4. extract and count significant parts of speech
5. define weight of the goal

```python
only_name['goal_name'] = only_name['goal_name'].apply(lambda x: str(x).lower())
```

```python
specials_to_remove = [
    '.', '"', "'", '?', '(', ')', '`',
]

def remove_special(x):
    for special in specials_to_remove:
        if special in x:
            x =  x.replace(special, '').strip()
        else:
            pass
    return x

specials_to_replace = [
    '-', '\\', '/', ','
]

def replace_special(x):
    for special in specials_to_replace:
        if special in x:
            x =  x.replace(special, ' ').strip()
        else:
            pass
    return x
```

```python
only_name['goal_name'] = only_name['goal_name'].apply(lambda x: remove_special(x))
only_name['goal_name'] = only_name['goal_name'].apply(lambda x: replace_special(x))
```

```python
def create_list_of_words(x):
    return x.split(' ')
```

```python
only_name['goal_name_LoW'] = only_name['goal_name'].apply(lambda x: create_list_of_words(x))
```

```python
morph = pymorphy2.MorphAnalyzer()
```

```python
key_pos = ['NOUN', 'VERB', 'NUMR', 'ADJF', 'ADJS', 'INFN'] # is it a good idea to leave only verbs and numrs?
```

```python
def extract_pos(x):
    poses = []
    for word in x:
        if word.isdigit() == True:
            poses.append(word)
        else:
            p = morph.parse(word)[0]
            normal_form = p.normal_form
            pos = p.tag
            stop = 0
            for s_pos in key_pos:
                if s_pos in pos:
                    poses.append(normal_form)
                else:
                    continue
    return ' '.join(poses)
```

```python
only_name['goal_name_key_pos'] = only_name['goal_name_LoW'].apply(lambda x: extract_pos(x))
```

```python
from collections import Counter
import operator
```

```python
def create_weights_bow(array):
    bow = []
    counter = Counter()
    list_ = array.tolist()
    for text in list_:
        list_of_words = text.split(' ')
        for word in list_of_words:
            if word != '' and word != 'nan':
                bow.append(word)
    counter = Counter(bow)
    sorted_bow = sorted(counter.items(), key=operator.itemgetter(1), reverse=True)
    # create weights
    weights = dict()
    for pair in sorted_bow:
        weights[pair[0]] = round(1/pair[1], 4)
    return weights
```

```python
goal_weights = create_weights_bow(only_name['goal_name_key_pos'])
```

```python
goal_weights
```

```python
def weigh_goal(x):
    goal_weight = int()
    words = x.split(' ')
    for word in words:
        try:
            goal_weight += goal_weights[word]
        except KeyError: # ''
            pass
    return goal_weight
```

```python
only_name['goal_weight'] = only_name['goal_name_key_pos'].apply(lambda x: weigh_goal(x))
```

New features:

1. N words original
2. N letters original
3. Average word length
4. N of verbs
5. N of nouns
6. N of numerics
7. N of adjectives
8. N of digits

```python
def word_counter(x):
    return len(x)
```

```python
only_name['goal_words'] = only_name['goal_name_LoW'].apply(lambda x: word_counter(x))
```

```python
def letters_counter(x):
    counter = int()
    for word in x:
        counter += len(word)
    return counter
```

```python
only_name['goal_letters'] = only_name['goal_name_LoW'].apply(lambda x: letters_counter(x))
```

```python
#word len
only_name['goal_aver_word_len'] = round(only_name['goal_letters'].div(only_name['goal_words']), 2)
```

```python
def pos_counter(x, pos_to_comp):
    pos_counter = int()
    for word in x:
        p = morph.parse(word)[0]
        pos = p.tag
        for pos_ in pos_to_comp:
            if pos_ in pos:
                pos_counter += 1
            else:
                pass
    return pos_counter
```

```python
only_name['goal_verbs_counter'] = only_name['goal_name_LoW'].apply(lambda x: pos_counter(x, ['VERB', 'INFN']))
only_name['goal_nouns_counter'] = only_name['goal_name_LoW'].apply(lambda x: pos_counter(x, ['NOUN']))
only_name['goal_numr_counter'] = only_name['goal_name_LoW'].apply(lambda x: pos_counter(x, ['NUMR']))
only_name['goal_adj_counter'] = only_name['goal_name_LoW'].apply(lambda x: pos_counter(x, ['ADJF', 'ADJS']))
```

```python
def digit_counter(x):
    digit_counter = int()
    for word in x:
        if word.isdigit() == True:
            digit_counter += 1
    return digit_counter
```

```python
only_name['goal_digit_counter'] = only_name['goal_name'].apply(lambda x: digit_counter(x))
```

```python
only_name.drop(columns=['goal_name_LoW', 'goal_letters'], inplace=True)
```

Default categories are obvious from the table above, so let's create a list of default values

```python
default_types = [
    'Получить знания в новой области',
    'Приобрести новые навыки',
    'Развить имеющиеся знания и навыки',
    'Включиться в новый проект/деятельность',
    'Получить официальное образование (с дипломом)',
    'Включиться в сообщество, наработать новые связи',
    'Изменить привычки'
]
```

```python
def create_default_type(x):
    for type_ in default_types:
        if type_ in x:
            return x
        else:
            pass
    return 'Свой вариант'
```

Will lower and remove descriptive info from goal domain (+ unstack stacked domains)

```python
def unstack_remove_descr(x):
    goal_domain_split = str(x).split(')')
    to_join = str()
    for domain_split in goal_domain_split:
        domain = domain_split.split('(')[0].lstrip().rstrip()
        to_join += domain + ' '
    return to_join.strip()
```
